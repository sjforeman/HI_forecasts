## Plotting forecasts for $z$-independent cosmological parameters

```python
%load_ext autoreload
%autoreload 2
```

```python
# import revelant packages
from FishLSS.headers import *

import plotting_helper_functions as ph
```

```python
forecast_dir = "../output/forecasts/"

noises = ["chordnoise", "cvnoise", "nonoise", "skyonly"]

surveys = [
    "puma_32k_opt", 
    "puma_32k_pess", 
    "puma_5k_opt",
    "puma_5k_pess",
    "chord_opt",
    "chord_pess",
]

options = [
    "no_sub_dknldDinv0.3",
    "no_sub_knl0.4",
    "no_sub_knlZel",
    "sub_dknldDinv0.3",
    "sub_knl0.4",
    "sub_knlZel",
]

param_basis = [
    "LCDM_parameters_Tbfree_noCMB_noLensing",
    "LCDM_parameters_Tbfree_Planck_SOCMB_noLensing",
    "LCDM_parameters_Tbfree_Planck_SOCMB_SOLensing",
    "LCDMplusMnu_parameters_Tbfree_noCMB_noLensing",
    "LCDMplusMnu_parameters_Tbfree_Planck_SOCMB_noLensing",
    "LCDMplusMnu_parameters_Tbfree_Planck_SOCMB_SOLensing",
    "LCDMplusNeff_parameters_Tbfree_noCMB_noLensing",
    "LCDMplusNeff_parameters_Tbfree_Planck_SOCMB_noLensing",
    "LCDMplusNeff_parameters_Tbfree_Planck_SOCMB_SOLensing",
    "LCDMplusOmegaK_parameters_Tbfree_noCMB_noLensing",
    "LCDMplusOmegaK_parameters_Tbfree_Planck_SOCMB_noLensing",
    "LCDMplusOmegaK_parameters_Tbfree_Planck_SOCMB_SOLensing",
]

noLSS_param_basis = [
    "LCDM_parameters_Tbfree_Planck_SOCMB_noLensing_noLSS",
    "LCDM_parameters_Tbfree_Planck_SOCMB_SOLensing_noLSS",
    "LCDMplusMnu_parameters_Tbfree_Planck_SOCMB_noLensing_noLSS",
    "LCDMplusMnu_parameters_Tbfree_Planck_SOCMB_SOLensing_noLSS",
    "LCDMplusNeff_parameters_Tbfree_Planck_SOCMB_noLensing_noLSS",
    "LCDMplusNeff_parameters_Tbfree_Planck_SOCMB_SOLensing_noLSS",
    "LCDMplusOmegaK_parameters_Tbfree_Planck_SOCMB_noLensing_noLSS",
    "LCDMplusOmegaK_parameters_Tbfree_Planck_SOCMB_SOLensing_noLSS",
]
```

```python
parameter_output = {}

for noise in noises:
    for option in options:
        for survey in surveys:
            try:
                for basis in param_basis:
                    with open(os.path.join(forecast_dir, noise, option, survey, f"{basis}.json")) as f:
                        parameter_output[noise, survey, option, basis] = json.load(f)
            except:
                print(f"Not found: {noise}/{option}/{survey}")
```

```python
for basis in noLSS_param_basis:
    with open(f"{forecast_dir}/cmb_plus_lensing/{basis}.json") as f:
        parameter_output[basis] = json.load(f)
```
