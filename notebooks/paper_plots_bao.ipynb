## Plotting BAO forecasts

```python
%load_ext autoreload
%autoreload 2
```

```python
from FishLSS.headers import *

import plotting_helper_functions as ph
```

```python
forecast_dir = "../output/forecasts/"

noises = ["chordnoise", "cvnoise", "nonoise", "skyonly"]

surveys = [
    "puma_32k_opt", 
    "puma_32k_pess", 
    "puma_5k_opt",
    "puma_5k_pess",
    "chord_opt",
    "chord_pess",
]

options = [
    "no_sub_dknldDinv0.3",
    "no_sub_knl0.4",
    "no_sub_knlZel",
    "sub_dknldDinv0.3",
    "sub_knl0.4",
    "sub_knlZel",
]
```

```python
rec_BAO_output = {}

for noise in noises:
    for option in options:
        for survey in surveys:
            try:
                with open(os.path.join(forecast_dir, noise, option, survey, "bao.json")) as f:
                    rec_BAO_output[noise, survey, option] = json.load(f)
            except:
                print(f"Not found: {noise}/{option}/{survey}")
```

## Plots

### $\alpha_\perp$

```python
ph.plot_AP(
    rec_BAO_output,
    "chordnoise",
    "onesigma_aperp", 
    r"$\sigma_{\alpha_\perp} [\%]$", 
    r"$\sigma_{\alpha_\perp}$ ratio", 
    ['dknldDinv0.3', 'knl0.4'],
    [r"$k_{\rm NL}(z) = 0.3D(z)^{-1}\, h\,{\rm Mpc}^{-1}$", r"$k_{\rm NL}(z) = 0.4h\,{\rm Mpc}^{-1}$"],
    ylim_abs=[0, 4], 
    ylim_ratio=[0.4, 1.01], 
    lw=1.5, 
    fill_alpha=0.3, 
    abs_dtick=0.5, 
    ratio_dtick=0.1,
    legend_loc='upper left'
)
plt.savefig("../plots/paper/bao_aperp_chordnoise.pdf")
```
