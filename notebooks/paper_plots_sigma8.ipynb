## Plotting $\sigma_8$ forecasts

```python
%load_ext autoreload
%autoreload 2
```

```python
from FishLSS.headers import *

import plotting_helper_functions as ph
```

```python
forecast_dir = "../output/forecasts/"

noises = ["chordnoise", "cvnoise", "nonoise", "skyonly"]

surveys = [
    "puma_32k_opt", 
    "puma_32k_pess", 
    "puma_5k_opt",
    "puma_5k_pess",
    "chord_opt",
    "chord_pess",
]

options = [
    "no_sub_dknldDinv0.3",
    "no_sub_knl0.4",
    "no_sub_knlZel",
    "sub_dknldDinv0.3",
    "sub_knl0.4",
    "sub_knlZel",
]
```

```python
sigma8_output = {}

for noise in noises:
    for option in options:
        for survey in surveys:
            try:
                with open(os.path.join(forecast_dir, noise, option, survey, "sigma8_fixedshape_Tbfree.json")) as f:
                    sigma8_output[noise, survey, option, "freeTb"] = json.load(f)
                with open(os.path.join(forecast_dir, noise, option, survey, "sigma8_fixedshape_Tbfixed.json")) as f:
                    sigma8_output[noise, survey, option, "fixedTb"] = json.load(f)
            except:
                print(f"Not found: {noise}/{option}/{survey}")
```

## Plots

```python
ph.plot_sigma8(
    sigma8_output, 
    "chordnoise",
    ['dknldDinv0.3', 'knl0.4'],
    [r"$k_{\rm NL}(z) = 0.3D(z)^{-1}\, h\,{\rm Mpc}^{-1}$", r"$k_{\rm NL}(z) = 0.4h\,{\rm Mpc}^{-1}$"],
    ylim_abs=[0, 50], 
    ylim_ratio=[0.4, 1.03], 
    lw=1.5, 
    fill_alpha=0.3, 
    abs_dtick=10, 
    ratio_dtick=0.2,
    legend_loc='upper left',
    figsize=(9, 3.7),
    Tb="freeTb"
)
plt.savefig("../plots/paper/sigma8_fixedshape_freeTb_chordnoise.pdf")
```
